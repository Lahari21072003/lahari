```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("country_wise_latest.csv")
```

```python
top_20 = df.sort_values('Confirmed', ascending=False).head(20)

fig = px.bar(top_20, x='Country/Region', y='Confirmed', color='Country/Region', # Changed 'Country' to 'Country/Region' for x and color
             title='Top 20 Countries by Confirmed COVID-19 Cases')

fig.show()
```

```python
import plotly.graph_objects as go

top10 = df.sort_values("Confirmed", ascending=False).head(10)

fig = go.Figure()

fig.add_trace(go.Scatter(x=top10["Country/Region"], y=top10["Confirmed"],
                         mode='lines+markers', name='Confirmed'))
fig.add_trace(go.Scatter(x=top10["Country/Region"], y=top10["Deaths"],
                         mode='lines+markers', name='Deaths'))
fig.add_trace(go.Scatter(x=top10["Country/Region"], y=top10["Recovered"],
                         mode='lines+markers', name='Recovered'))
fig.add_trace(go.Scatter(x=top10["Country/Region"], y=top10["Active"],
                         mode='lines+markers', name='Active'))

fig.update_layout(title="COVID-19 Cases in Top 10 Countries",
                  xaxis_title="Country",
                  yaxis_title="Number of Cases",
                  template="plotly_dark")

fig.show()
```

# 3d visualization

```python
import plotly.graph_objects as go

top20 = df.sort_values("Confirmed", ascending=False).head(20)

fig = go.Figure(data=[go.Scatter3d(
    x=top20['Confirmed'],
    y=top20['Deaths'],
    z=top20['Recovered'],
    mode='markers+text',
    text=top20['Country/Region'],
    marker=dict(
        size=8,
        color=top20['Active'],  # Use Active cases for color
        colorscale='Viridis',
        opacity=0.8
    )
)])

fig.update_layout(
    title='3D Visualization of COVID-19 Data by Country',
    scene=dict(
        xaxis_title='Confirmed Cases',
        yaxis_title='Deaths',
        zaxis_title='Recovered'
    ),
    template="plotly_dark"
)

fig.show()
```

# Box plot

```python
df = pd.read_csv('country_wise_latest.csv')

box_data = df[['Confirmed', 'Deaths', 'Recovered', 'Active']]

melted = box_data.melt(var_name='Category', value_name='Count')


fig = px.box(melted, x='Category', y='Count', color='Category',
             title='COVID-19 Case Distribution by Category (All Countries)',
             points='all',
             template='plotly_dark')

fig.show()
```

```python
df = pd.read_csv("country_wise_latest.csv")

df['Increase Group'] = pd.cut(df['1 week % increase'],
                              bins=[-1, 5, 10, 20, 50, 100],
                              labels=['<5%', '5–10%', '10–20%', '20–50%', '>50%'])

df = df.dropna(subset=['Increase Group'])

fig = px.scatter(df, x="Confirmed", y="Deaths",
                 animation_frame="Increase Group", animation_group="Country/Region",
                 size="Recovered", color="WHO Region", hover_name="Country/Region",
                 facet_col="WHO Region", log_x=True, size_max=45,
                 range_x=[100, 1000000], range_y=[0, 100000],
                 title="Simulated COVID-19 Animation by Weekly Case Increase")

fig.show()
```

```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("country_wise_latest.csv")

fig = px.scatter_geo(
    df,
    locations="Country/Region",      # Country name column
    locationmode="country names",    # Match based on full country names
    color="Confirmed",               # Color by number of confirmed cases
    size="Deaths",                   # Bubble size based on deaths
    hover_name="Country/Region",     # Show country on hover
    projection="natural earth",
    title="COVID-19 Global Spread: Confirmed vs Deaths"
)

fig.show()
```

# IMAGES AND HEAT MAPS

```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("country_wise_latest.csv")

heatmap_df = df.set_index("Country/Region")[["Confirmed", "Deaths", "Recovered", "Active"]]

heatmap_df = heatmap_df.sort_values(by="Confirmed", ascending=False).head(30)

# Create the heatmap
fig = px.imshow(
    heatmap_df,
    labels=dict(x="Metric", y="Country", color="Value"),
    x=heatmap_df.columns,
    y=heatmap_df.index,
    color_continuous_scale="YlOrRd",
    title="COVID-19 Heatmap: Countries vs Metrics"
)

fig.show()
```

# COVID-19 Global Dataset Summary

The COVID-19 Global Dataset offers a comprehensive snapshot of the pandemic's impact across various countries, encompassing key metrics like confirmed cases, deaths, recoveries, and active cases. It provides not only total figures but also recent changes in these metrics, allowing for dynamic tracking of the virus's spread and impact. The dataset includes region classifications based on WHO standards, offering insights into global and regional trends. With information on total tests conducted, critical cases, and population sizes, it enables the analysis of per-capita metrics and testing efficacy. This dataset allows for in-depth trend analysis, helping to compare different countries and regions in terms of case rates, recovery rates, and mortality rates. Additionally, it can be leveraged for understanding the effectiveness of government policies, healthcare responses, and the global distribution of resources. Through visualizations like global heatmaps and country-level statistics, this data serves as a crucial tool for monitoring the ongoing impact of the pandemic and guiding future responses.
