# **multivariate analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("country_wise_latest.csv")


num_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active']


sns.pairplot(df[num_cols])
plt.suptitle("Pairplot of COVID Metrics", y=1.02)
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

 Scatter Plot  Deaths vs Confirmed

```python
sns.lmplot(x="Confirmed", y="Deaths", data=df, height=6, aspect=1.2)
```

Pie Chart  Proportion of Total Cases


```python
labels = ['Active', 'Recovered', 'Deaths']
sizes = [df['Active'].sum(), df['Recovered'].sum(), df['Deaths'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title("Global Case Distribution")
plt.show()
```

Bar Plot  Top 10 Countries by Confirmed Cases

```python
top10 = df.sort_values("Confirmed", ascending=False).head(10)
sns.barplot(data=top10, x="Confirmed", y="Country/Region", palette="Reds_r")
```
